```python
devliveries_df[['match_id','Balls_score']].groupby('match_id').cumsum()['Balls_score']
```

```python
#new column which has target
devliveries_df['target_left']=(devliveries_df['total_runs']+1)-devliveries_df['Score']
```

```python
#1 over 6 balls
#ipl has 20 over match 20*6=120
devliveries_df['Remaining Balls']=(120-((devliveries_df['over']-1)*6+devliveries_df['ball']))
```

```python
devliveries_df
```

```python
devliveries_df['player_dismissed']=devliveries_df['player_dismissed'].fillna('0')
```

```python
devliveries_df['player_dismissed']
```

```python
devliveries_df['player_dismissed']=devliveries_df['player_dismissed'].apply(lambda x:x if x=='0' else '1').astype('int64')
```

```python
devliveries_df
```

```python
devliveries_df['Wickets']=devliveries_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
devliveries_df
```

```python
devliveries_df['Wickets']=10-devliveries_df['Wickets']
```

```python
devliveries_df
```

Current Runrate

$$Current \ run \ rate = \frac{Runs \ scored}{Number \ of \ Overs \ completed}$$

```python
devliveries_df['crr']=(devliveries_df['Score'])*6/(120-devliveries_df['Remaining Balls'])
```

```python
devliveries_df
```

Required Runrate

$$Required \ run \ rate = \frac{Remaining \ score \ or \ target }{Number \ of \ Overs \ left}$$

```python
devliveries_df['rrr']=(devliveries_df['target_left'])*6/(devliveries_df['Remaining Balls'])
```

```python
devliveries_df
```

```python
#creating column result where 1 means won 0 means lost
def result(row):
    if row['batting_team']==row['winner']:
        return 1
    else:
        return 0
    
```

```python
devliveries_df['result']=devliveries_df.apply(result,axis=1)
```

```python
Model_df=devliveries_df[[ 'batting_team','bowling_team','city','Score', 'Wickets', 'Remaining Balls','target_left','crr', 'rrr','result']]
```

```python
Model_df=Model_df[Model_df['Remaining Balls']!=0]
```

```python
Model_df=Model_df.sample(Model_df.shape[0])
```

```python
#Splitting the data into features and targets
X=Model_df.iloc[:,:-1]
Y=Model_df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
#Spliting the data to train and test
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_test
```

```python
Y_test
```

```python
Y_train
```

```python
#label encoding
#one hot encoder
#transform non numerical to numerical

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])],remainder='passthrough')
```

```python
trf
```

```python
#building the model
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_prediction=pipe.predict(X_test)
```

```python
Y_prediction
```

```python
#Accuracy score 
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_prediction)
```
